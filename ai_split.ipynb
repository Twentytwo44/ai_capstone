# Test accuracy use vector search

```python
test_data = {
   
    "Python, Java, Cloud Computing": ["Innovatech", "Innovatech", "TechTonic", "TechTonic", "TechTonic", "TechTonic", "TechTonic", "TechTonic", "CloudBuilders", "ServerCrafters"],
    "UI/UX Design": ["DesignEmporium", "WebCreators", "CraftyStudios", "DesignStudio", "InteractiveDesign", "DesignGenius", "WebWorks", "UI/UX Solutions", "FrontendFocus", "UXPrototype"],
    "Ai Machine Learning": ["AI Dynamics", "AI Dynamics", "AI Dynamics", "AI Dynamics", "AI Dynamics", "AI Dynamics", "AI Innovations", "AI Innovations", "AI Innovations", "AI Innovations"],
    "Web Development": ["TechWave", "TechWave", "WebInteract", "CloudSolutions", "CloudSolutions", "CloudSolutions", "CloudSolutions", "CloudSolutions", "CloudSolutions", "TechTonic"],
    "Mobile Application": ["InterfaceForge", "UXVision", "AppUX", "MobileMinds", "MobileMinds", "MobileMinds", "MobileMinds", "MobileMinds", "MobileMinds", "MobileUX"]
}
def measure_accuracy(test_data, embedded_data, embedded_test, top_k=10):
    correct_predictions = 0
    total_predictions = 0

    for query, recommendations in test_data.items():
        query_embedding = embedded_test[test_text.index(query)]
        similarity_scores = cosine_similarity([query_embedding], embedded_data)[0]
        top_indices = similarity_scores.argsort()[-top_k:][::-1]
        top_recommendations = [Company['Name'].iloc[index] for index in top_indices]

        for recommendation in recommendations:
            if recommendation in top_recommendations:
                correct_predictions += 1
            total_predictions += 1

    accuracy = correct_predictions / total_predictions
    return accuracy

accuracy = measure_accuracy(test_data, embedded_data, embedded_test)
print("Accuracy:", accuracy)

```
